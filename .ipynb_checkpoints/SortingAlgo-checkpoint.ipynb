```python
# loading required libraries
import random
import timeit
```

```python
# generation a random dataset
start_rand_gen = timeit.timeit() # to start timer

start = input('please enter starting integer ' ) # start integer value of the list
stop = input('please enter stopping integer ') # end integer value of the list
length_iter = input('please enter the length ') # lnength desired for the list 

listrandom=[]  #initiatig an empty list

for i in range(int(length_iter)):                    # loop for generating the data iteration in the given range
    value=random.randint(int(start),int(stop))
    listrandom.append(value)                         # populating the empty list 

print(type(listrandom))                              # checking the list property
print(listrandom)                                    # printing the list

end_rand_gen = timeit.timeit() # to end timer
Time2gen_rand_list = end_rand_gen - start_rand_gen
print("\n time to generate the random list {} seconds".format(Time2gen_rand_list))
```

```python
# Sorting Algorithm
## we will be employing QuickSort, wherein we will select an arbitary Data as pivot and will sort data lower sider and higher side (called partitions) 
## with its reference recursively. This algorithm is evaluated in linear time space and is done in-place, which makes is quick with  yields average time 
## complexity of O(n log n), with low overhead 

# creating sorting function 
def sort_Algo(x):
    lower_side = []
    pivot_List = []
    higher_side = []
    if len(x) <= 1:
        return x
    else:
        pivot = x[0] # picking first element as pivot
        for i in x:
            if i < pivot:
                lower_side.append(i)
            elif i > pivot:
                higher_side.append(i)
            else:
                pivot_List.append(i)
        lower_side = sort_Algo(lower_side)
        higher_side = sort_Algo(higher_side)
        sorted_array = lower_side + pivot_List + higher_side
        return sorted_array
```

```python
start_sort_timer = timeit.timeit() # to start timer

sorted_list = sort_Algo(listrandom)
print(sorted_list)
print("\n size of sorted list {}".format(len(sorted_list)))

end_sort_timer = timeit.timeit() # to end timer
Time2gen_sorting_list = end_sort_timer - start_sort_timer
print("\n time to sort the list {} seconds".format(Time2gen_sorting_list))
sorted_list = []    # initiating and empty string to store sorted data 
```

```python
## in this case space complexity is O(log n) and time complexity is  O(n log n) as partitions are nearly than n/2.
```
